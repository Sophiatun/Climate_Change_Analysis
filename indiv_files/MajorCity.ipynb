```python
cities_grouped = major_cities_df.groupby("City")
cities_grouped.head()
```

```python
# Find the earliest date all countries have in common

earliest_date = cities_grouped["dt"].min().max()
# earliest_date = dt.datetime(earliest_date.year, earliest_date.month, earliest_date.day).date()
#print(earliest_date)
```

```python
latest_date = cities_grouped["dt"].max().min()
# latest_date = dt.datetime(latest_date.year, latest_date.month, latest_date.day).date()
# print(latest_date)
```

```python
date_range = (f"Our date range is {earliest_date} to {latest_date}")
print(date_range)
```

```python
cleaned_cities = major_cities_df[major_cities_df['dt']>='1950-02-01']
cleaned_cities.head()
```

```python
final_df = cleaned_cities[["dt", "AverageTemperature", "City", "Country"]]
final_df.head()
```

```python
final_df.rename(columns={'AverageTemperature': "Average Temperature",
                         'dt': "Date"}, inplace=True)
final_df.head()
```

```python
final_df.to_csv("MajorCities.csv")
```
